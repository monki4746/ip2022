```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img1 = cv2.imread('dsu1.jpg')
rows,cols,ch = img1.shape



pts1 = np.float32([[502,291],[758,310],[506,526],[761,506]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])

M = cv2.getPerspectiveTransform(pts1,pts2)

dst = cv2.warpPerspective(img1,M,(300,300))

plt.figure(figsize=(10,10), facecolor="white")

plt.subplot(221),plt.imshow(img1),plt.title('Input')
plt.subplot(222),plt.imshow(dst),plt.title('Output')

plt.show()
```
